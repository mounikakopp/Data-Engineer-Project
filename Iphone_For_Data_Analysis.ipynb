```python
## Fuction to extract color from product name
import re
def extract_color(product_name):
    match = re.search(r'\(([^,]+),', product_name)
    if match:
        return match.group(1)
    else:
        return None
```

```python
df['Color'] = df['Product Name'].apply(extract_color)
```

```python
df.head()
```

```python
# Products which are priced less than 50000 and are in black color
df[(df.Mrp <= 50000) & (df.Color == "Black")]
```
